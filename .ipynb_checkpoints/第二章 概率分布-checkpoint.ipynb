# 一、离散分布

## 1.1 二项分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom as B
%matplotlib inline

rv = B(10, 0.5)
x = np.arange(0, 11, 1)
y = rv.pmf(x)

print y
plt.bar(x, y, width = 0.6, color = 'grey')
plt.show()
```

## 1.2 几何分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom as G
%matplotlib inline

rv = G(0.2)
x = np.arange(0, 11, 1)
y = rv.pmf(x)

print y
plt.bar(x, y, width = 0.6, color = 'grey')
plt.show()
```

## 1.3 泊松分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson as Pie
%matplotlib inline

rv = Pie(4.5)
x = np.arange(0, 11, 1)
y = rv.pmf(x)

print y
plt.bar(x, y, width = 0.6, color = 'grey')
plt.show()
```

# 二、连续分布

## 2.1 指数分布

###  2.1.1 指数分布的概率密度函数

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon as E
%matplotlib inline

rv1 = E(scale = 1.5)
rv2 = E(scale = 1.0)
rv3 = E(scale = 0.5)
x = np.linspace(0, 20, 100)

plt.plot(x, rv1.pdf(x), color = 'green')
plt.plot(x, rv2.pdf(x), color = 'blue')
plt.plot(x, rv3.pdf(x), color = 'red')
plt.show()
```

### 2.1.2 指数分布的分布函数

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon as E
%matplotlib inline

rv = E(scale = 1.0)
x = np.linspace(0, 5, 100)

plt.plot(x, rv.pdf(x), color = 'blue')
plt.plot(x, rv.cdf(x), color = 'red')
plt.show()
```

### 2.1.3 模拟产生指数分布随机变量

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon as E
def exp(lam):
    p = random.random()
    return -math.log(1-p)/lam
x1 = []
for i in range(100):
    x1.append(exp(1))
x1 = sorted(x1)
y1 = np.linspace(0, 1, 100)
plt.plot(x1, y1, color = 'blue')

rv = E(scale = 1)
x2 = np.linspace(0, 5, 100)
plt.plot(x2, rv.cdf(x2), color = 'red')
plt.show()
```

## 2.2 正态分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm as N
%matplotlib inline

rv1 = N(loc = 0, scale = 1)
rv2 = N(loc = -5, scale = 1)
rv3 = N(loc = 0, scale = 3)
x = np.linspace(-10, 10, 100)

plt.plot(x, rv1.pdf(x), color = 'green')
plt.plot(x, rv2.pdf(x), color = 'blue')
plt.plot(x, rv3.pdf(x), color = 'red')
plt.show()
```
